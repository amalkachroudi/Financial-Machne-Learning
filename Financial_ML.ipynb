** one month changes of M1 **

```python
data_M1 = pd.read_csv('2000\M1.csv',sep=',')
data_M1.head(5)
```

```python
n1=data_M1.shape[0]
data_M1['change']=0.00
for i in range(1,n1):
    data_M1['change'][i]=(data_M1['M1'][i]-data_M1['M1'][i-1])/data_M1['M1'][i-1]
data_M1.head()
```

```python
import time
data_M1['month_change']='NaN'
S=data_M1['change'][0]
c=1
for i in range(0,n1-1):
    if time.strptime(data_M1['DATE'][i],"%Y-%m-%d")[1]==time.strptime(data_M1['DATE'][i+1],"%Y-%m-%d")[1]:
        S=S+data_M1['change'][i+1]
        c=c+1
    else:
        data_M1['month_change'][i]=S/c
        c=1
        S=data_M1['change'][i+1]
        
```

```python
data_M1['monthly_change']=0.00
a=data_M1['M1'][0]
for i in range(0,n1-1):
    if time.strptime(data_M1['DATE'][i],"%Y-%m-%d")[1]!=time.strptime(data_M1['DATE'][i+1],"%Y-%m-%d")[1]:
        data_M1['monthly_change'][i]=(data_M1['M1'][i]-a)/a
        a=data_M1['M1'][i+1]
        
```

```python
data_M1.head()
```

```python
df=pd.DataFrame(columns=['DATE','monthly_change'])
df['DATE']=data_M1['DATE']
df['monthly_change']=data_M1['monthly_change']
```

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
def bscatter(X,Y,k):
    Xnom, thresh=pd.qcut(X, k, retbins=True)
    tp = pd.DataFrame({'Xnom': Xnom, 'X': X, 'Y':Y})
    GMn  = tp.groupby('Xnom').mean()
    SeMn = tp.groupby('Xnom').std()/ np.sqrt(tp.groupby('Xnom').count())
    Xq = GMn.X
    Yq = GMn.Y
    XSe = SeMn.X
    YSe = SeMn.Y
    return(Xq,XSe,Yq,YSe)
```

**Assumptions of Linear Regression**

* Linear relationship
* Multivariate normality
* No or little multicollinearity
* No autocorrelation of residuals
* Homoscedasticity of residuals or equal variance
* Normality of residuals
* The X variables and residuals are uncorrelated
* The mean of residuals is zero
* The number of observations must be greater than number of Xs
